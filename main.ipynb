```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data=df, x='year', y='price',hue="transmission")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=df, x=df['mileage'], y=df['price'], hue="transmission", style="fuelType", s=100) # s=100 nuqtalar hajmi
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1) # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 1-grafik
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish


plt.subplot(2,1,2) # # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 2-grafik
# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish

plt.show()
```
